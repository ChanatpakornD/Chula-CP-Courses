# Signals
## Lab 01 - Intro to signal processing
* Plot $x(t) = 2e^{j2 \pi t}$

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')
t = np.arange(-1*np.pi,1*np.pi,0.01)
complex_expo = 2*np.e**(1j*2*np.pi*t)
real = np.real(complex_expo)
imag = np.imag(complex_expo)


ax.plot3D(t, real, imag)
ax.set_xlabel('t ')
ax.set_ylabel('real ')
ax.set_zlabel('imag ')

plt.show()

plt.plot(t,real,label="real")
plt.plot(t,imag,label="imag")
plt.legend(loc="upper left")
plt.show()
```

Periodic or aperiodic?
* $x_1(t) = sin(\frac{2 \pi  }{ 3}t) + sin(3 \pi t)  $
* $x_2(t) = sin(\frac{2 \pi  }{ 3}t) + sin(3 t)  $

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-10,10,0.01)
x1 = np.sin(2 * np.pi/3 * t) + np.sin(3 * np.pi * t)
x2 = np.sin(2 * np.pi/3 * t) + np.sin(3 * t)


plt.plot(t,x1)
plt.show()
plt.plot(t,x2)
plt.show()
```
