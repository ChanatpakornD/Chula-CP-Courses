## Problem 1
Sketch the following signals

1.
$x(t) = \sin \left (\frac{\pi}{4}t + 20^\circ \right)$


2.
$x(t) = \begin{cases}t+2 , t \leq -2 \\ 0 , -2 \leq t \leq 2\\ t-2 , t \geq 2 \end{cases}$

3.
$x(t) = 2e^{-t}, 0 \leq t < 1$ and $ x(t+1) = x(t) $ for all $t$

4.
$x(t) = u(t) + 5u(t-1) -2u(t-2)$

5.
$x(t) = r(t) - r(t-1) - u(t-2)$




```
def u(x):
  return np.where(x > 0, 1, 0)

def r(x):
  return np.where(x >= 0, x, 0)

t = np.linspace(-10,10,1000)

p1 = np.sin(np.pi / 4 * t + np.pi / 9)

p2 = np.piecewise(t,
                  [t <= -2, (t >= -2) & (t <= 2), t >= 2],
                  [lambda t : t+2 , lambda t : 0 , lambda t : t-2])

p3 = 2 * np.exp(-1 * (t%1))

p4 = u(t) + 5*u(t-1) - 2*u(t-2)

p5 = r(t) - r(t-1) - u(t-2)

plt.figure(figsize = (6,4))
plt.plot(t, p1)
plt.show()
plt.figure(figsize = (6,4))
plt.plot(t, p2)
plt.show()
plt.figure(figsize = (6,4))
plt.plot(t, p3)
plt.show()
plt.figure(figsize = (6,4))
plt.plot(t, p4)
plt.show()
plt.figure(figsize = (6,4))
plt.plot(t, p5)
plt.show()
```
